```python
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Grouper
from pandas import DataFrame
```

```python
serie_o = pd.read_csv("/home/jonathan/tesis/1er_avance/timeseries_d.csv", sep=',',index_col=0, parse_dates=True, usecols=["ds","2807918","2807913","2807903","2807901","2807907","2807906","2807905"])
serie_o.shape
serie_o.head(5)
```

```python
serie_o['Fecha'] = serie_o.index.date
serie_o['Año'] = serie_o.index.year                                
serie_o['Mes'] = serie_o.index.month                               
serie_o['ID_Dia'] = serie_o.index.dayofweek
serie_o['Hora'] = serie_o.index.hour
serie_o.head(30)
#serie_o.sample(8, random_state=0)  
```

```python
data1=serie_o.loc['2020-02-14':'2020-03-15']
data2=serie_o.loc['2020-03-15':'2020-06-30']
data3=serie_o.loc['2020-07-01':'2020-09-30']
data4=serie_o.loc['2020-10-01':'2020-12-31']
```

```python
table1_1 = pd.pivot_table(data1, values='2807901', index=['Hora'],
                    columns=['Fecha'],fill_value=0)
table1_1.head(5)
```

```python
table2_1 = pd.pivot_table(data1, values='2807901', index=['Fecha'],
                    columns=['Hora'],fill_value=0)
table2_1.head(5)
```

```python
data = table2_1.iloc[:,0:1]
data
```

```python
sb.set(rc = {'figure.figsize':(15,8)})
fig, ax = plt.subplots()          
fig.subplots_adjust(right=0.75)   

ax1 = ax.twinx()
ax1.spines.left.set_position(("axes", 0.5))
data = table2_1.iloc[:,0:1]

p1 = data.plot(marker='_',linestyle='none',color='g',label="Density",ax=ax)
ax1.yaxis.set_ticks([])

ax.set_xlabel("Hora")  
ax.set_ylabel("Número de Viajes")
ax1.set_ylabel("2020-02-14 : 2020-03-15", size=12)

p1.get_legend().remove()
```
